```python
df.drop(["Status",'unnamed1'], axis=1, inplace = True)
```

```python
df.info()
```

```python
df.dropna(inplace = True)
```

```python
pd.isnull(df).sum()
```

```python
df["Amount"] = df["Amount"].astype(int)
```

```python
df.info()
```

```python
df.describe()
```

```python
ax = sms.countplot(x = "Gender",data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(["Gender"] , as_index = False)["Amount"].sum().sort_values(by="Amount",ascending = False)
```

```python
sms.barplot(x= "Gender", y= "Amount",data = sales_gen)
```

```python
df.columns
```

```python
sms.countplot(data = df , x= "Age Group", hue = "Gender")
```

```python
ax = sms.countplot(data = df , x= "Age Group", hue = "Gender")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(["Age Group"] , as_index = False)["Amount"].sum().sort_values(by="Amount",ascending = False)
sms.barplot(x= "Age Group", y= "Amount",data = sales_age)
```

```python
sales_state = df.groupby(["State"] , as_index = False)["Orders"].sum().sort_values(by="Orders",ascending = False).head(10)
sms.set(rc ={"figure.figsize":(15,8)})
sms.barplot(x= "State", y= "Orders",data = sales_state)
```

```python
sales_state = df.groupby(["State"] , as_index = False)["Amount"].sum().sort_values(by="Amount",ascending = False).head(10)
sms.set(rc ={"figure.figsize":(15,8)})
sms.barplot(x= "State", y= "Amount",data = sales_state)
```

```python
ax1 = sms.countplot(x = "Marital_Status",data = df,hue = "Gender")
sms.set(rc ={"figure.figsize":(7,5)})
for bars in ax1.containers:
    ax1.bar_label(bars)
```
